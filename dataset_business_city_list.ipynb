### Cities list in LA area
data collection from https://en.wikipedia.org/wiki/List_of_cities_in_Los_Angeles_County,_California

```python
from bs4 import BeautifulSoup as BS
import requests
import pandas as pd
```

#### Attempt 1 


```python
web_url = 'https://en.wikipedia.org/wiki/List_of_cities_in_Los_Angeles_County,_California'
html_content = requests.get(web_url).text
soup = BS(html_content, 'html.parser')
```

```python
data = []
counter = 0
for row in soup.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) >= 3:
        city_link = cells[0].find('a')
        if city_link and 'California' in city_link.get('title', ''):
            city = city_link.get_text(strip=True)
            incorporation_date = cells[1].get_text(strip=True)
            population = cells[2].get_text(strip=True)
            data.append([city, incorporation_date, population])
```

```python
df = pd.DataFrame(data, columns=['City', 'Date_incorporated', 'Population'])
df['Date_incorporated'] = pd.to_datetime(df['Date_incorporated'])
df['Population'] = df['Population'].str.replace(',', '').astype(int)
df.info()
```

```python
df.to_csv('city_list.csv', index=False) 
```
